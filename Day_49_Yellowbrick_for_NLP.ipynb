```python
from sklearn.feature_extraction.text import TfidfVectorizer
from yellowbrick.text import TSNEVisualizer
from yellowbrick.datasets import load_hobbies
```

### Load Corpus

```python
corpus = load_hobbies()
corpus.labels
```

### Vectorize the Documents

```python
vec = TfidfVectorizer()
docs = vec.fit_transform(corpus.data)
labels= corpus.target
docs.shape
```

### Cluster Similar Documents

```python
# Euclidean Distance
tsne = TSNEVisualizer(size=(700,500),metrics = 'euclidean')
tsne.fit(docs,labels)
tsne.poof()
```

### Manhattan Distance

```python
#cityblock
tsne = TSNEVisualizer(metic = 'cityblock',size = (700,500))
tsne.fit(docs,labels)
tsne.poof()
```

### Bray Crutis Dissimilarity

```python
#braycurtis
tsne = TSNEVisualizer(metric='braycurtis',size=(700,500))
tsne.fit(docs,labels)
tsne.poof()
```

### Canberra Distance

```python
tsne = TSNEVisualizer(metric='canberra',size=(700,500))
tsne.fit(docs,labels)
tsne.poof()
```

### Cosine Distance

```python
tsne = TSNEVisualizer(metric='cosine',size=(700,500))
tsne.fit(docs,labels)
tsne.poof()
```
